### Requesting URLs
Most APIs are accessed using URL requests, much like requesting a page in a web browser. For our purposes, a URL is just a file or application that we can access using a web address (e.g., http://bc.edu). Python has features to help us make the requests.

The APIs we'll work with have URLs thst return data in raw formats such as `json` or `XML` files. These differ from the `html` files that we typically read in a web browswers. You learned about structured and unstructured data earlier. APIs return structured data; `html` files typically return unstructured data that is rendered for people to easily read.

### API example
We'll use GitHub a lot in our data science work, so let's use its API for our example. The GitHub API is accessible at `api.github.com.` This URL will return structured data files from Github.    

```python
import pandas as pd

# The `requests` library allows us to make URL requests and retrieve data from
# URLs
import requests
```

Now we can make make a request to the GitHub API to return information about a specific user. That user should be you...enter your GitHub name in the cell below and run it to make the API request.

```python
# request data from GitHub API about your account

# FILL IN YOUR GITHUB USER NAME BELOW
github_account_name = "FILL IN YOUR GITHUB USER NAME HERE"

assert github_account_name != "FILL IN YOUR GITHUB USER NAME HERE", f"Make sure to add your GitHub user name to the variable githib_account_name!"

# The GitHub API URL is https://api.github.com/users/YOUR_USER_NAME.
# Put your user name in the gihub_account_name variable named 
# github_account_name. We use the '+' operator below to add 
# the API URL with your user name

page = requests.get('https://api.github.com/users/' + github_account_name)

# the content we get back from `requests.get` is a json file
page.content
```

JSON data is highly structured, but it's not always easy for people to read. We can make this easier by loading the data into a pandas dataframe. To do this, we call `page.json()` to return the data as a Pyhton dictionary. We can make the dataframe from that dictionary.

```python
df = pd.DataFrame.from_dict(page.json(), orient='index')
df
```
